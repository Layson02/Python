```python
n = int(input())
a, b = 0, 1
while a <= n:
    print(a)
    a, b = b, a + b
```

```python
n = int(input())
divisores = 0
for i in range(1, n + 1):
    if n % i == 0:
        divisores += 1
if divisores == 2:
    print("Primo")
else:
    print("Não primo")
```
