# Trabalhando com tabela hash

A **tabela hash** é uma estrutura de dados que associa chaves a valores e permite acesso rápido aos dados através de um processo chamado **hashing**.

Essa estrutura oferece acesso e manipulação eficientes, possibilitando a recuperação e atualização de valores com complexidade média de tempo constante, ou O(1), na maioria dos casos. É ideal para situações onde precisamos buscar, adicionar ou remover dados de forma rápida e eficiente.

```python
class HashTable:
    def __init__(self, size):
        self.size = size
        self.table = [[] for _ in range(size)]

    def _hash_function(self, key):
        return key % self.size

    def insert(self, value):
        key = value % self.size
        index = self._hash_function(key)
        self.table[index].append(value)

    def get(self, key):
        index = self._hash_function(key)
        for val in self.table[index]:
            if val == key:
                return index
        raise KeyError(f"Key '{key}' not found")

    def remove(self, key):
        index = self._hash_function(key)
        for i, val in enumerate(self.table[index]):
            if val == key:
                del self.table[index][i]
                return
        raise KeyError(f"Key '{key}' not found")

    def imprimir(self):
        for index, bucket in enumerate(self.table):
            print(f"Posição {index}: {bucket}")
```

# Aplicação da estrutura de Tabela Hashing

```python
# Exemplo 01
hash_table = HashTable(5)  # Tamanho conforme requisitado

hash_table.insert(78)
hash_table.insert(60)
hash_table.insert(96)
hash_table.insert(59)
hash_table.insert(13)

hash_table.imprimir()
```

## Questão 01 - Verifique se a lista de valores apresenta colisão quando utilizamos uma tabela hash. Valore[120, 123, 145, 90, 39, 45, 23, 220]?

```python
# Seu código aqui

hash_table = HashTable(8)

hash_table.insert(120)
hash_table.insert(123)
hash_table.insert(145)
hash_table.insert(90)
hash_table.insert(39)
hash_table.insert(45)
hash_table.insert(23)
hash_table.insert(220)

print("\n--- Colisões Detectadas ---\n")
total_colisoes = 0

for i, colisoes in enumerate(hash_table.table):
    if len(colisoes) > 1:
        print(f"Posição {i} teve colisão: {colisoes}")

        total_colisoes += (len(colisoes) - 1)

print(f"\nTotal de colisões encontradas: {total_colisoes}\n")

hash_table.imprimir()
```

## Questão 03 - Verifique se a lista de caracteres apresenta colisão quando utilizamos uma tabela hash. Valore['U','N','I','E','S','P','F','A','C','U','L','D','A','D','E']?

```python
# Seu código aqui

nome='UNIESPFACULDADE'
hash_table = HashTable(len(nome))

for i in nome:
  hash_table.insert(ord(i))

print(f"\nQuantidade total de chaves inseridas: {len(nome)}")

print("\n--- Colisões Detectadas ---\n")
total_colisoes = 0

for i, colisoes in enumerate(hash_table.table):
    if len(colisoes) > 1:
        print(f"Posição {i} teve colisão: {colisoes}")

        total_colisoes += (len(colisoes) - 1)

print(f"\nTotal de colisões encontradas: {total_colisoes}\n")

hash_table.imprimir()
```

## Questão 04 - Dado os valores 2341, 4234, 2839, 430, 22, 397, 3920, uma tabela hash de tamanho 7, e função de hash h(x) = x % 7, mostre o resultante depois de inserir os valores na ordem dada.

## Questão 05 - Dada uma tabela hash de tamanho 17, se as chaves 2, 32, 43, 16, 77, 51, 1, 17, 42, 111 forem inseridas sequencialmente com a função de hash h(k) = k % 17, mostre o resultado depois de inserir os valores na ordem dada.

```python
# Seu código aqui

hash_table = HashTable(17)

chaves = [2, 32, 43, 16, 77, 51, 1, 17, 42, 111]

for chave in chaves:
    hash_table.insert(chave)

print(f"\nQuantidade total de chaves inseridas: {len(chaves)}")

print("\n--- Colisões Detectadas ---\n")
total_colisoes = 0

for i, colisoes in enumerate(hash_table.table):
    if len(colisoes) > 1:
        print(f"Posição {i} teve colisão: {colisoes}")

        total_colisoes += (len(colisoes) - 1)

print(f"\nTotal de colisões encontradas: {total_colisoes}\n")

hash_table.imprimir()
```

## Questão 06 - Desenhe uma tabela de hash resultante da introdução das chaves 12, 44, 13, 88, 23, 94, 11, 39, 20, 16 e 5, usando a função de hash h(k) = (2k + 5)%11.

```python
# Seu código aqui

hash_table = HashTable(11)
chaves = [12, 44, 13, 88, 23, 94, 11, 39, 20, 16, 5]

for k in chaves:
    hash_table.insert(2 * k + 5)

print(f"\nQuantidade total de chaves inseridas: {len(chaves)}")

print("\n--- Colisões Detectadas ---\n")
total_colisoes = 0

for i, colisoes in enumerate(hash_table.table):
    if len(colisoes) > 1:
        print(f"Posição {i} teve colisão: {colisoes}")

        total_colisoes += (len(colisoes) - 1)

print(f"\nTotal de colisões encontradas: {total_colisoes}\n")

hash_table.imprimir()
```

## Questão 07 (extra) - Dado os valores 2341, 4234, 2839, 430, 22, 397, 3920, construa uma tabela hash de tamanho 7, depois verifique se existe colisão e mostre a chave hash dos valores inseridos.

```python
# Seu código aqui

hash_table = HashTable(7)

valores = [2341, 4234, 2839, 430, 22, 397, 3920]

print("--- Mapeamento de Chaves Hash ---\n")
for v in valores:
    indice = v % 7 # Cálculo da chave hash
    print(f"Valor {v} -> Chave Hash: {indice}")
    hash_table.insert(v)

print(f"\nQuantidade total de chaves inseridas: {len(valores)}")

print("\n--- Colisões Detectadas ---\n")
total_colisoes = 0

for i, colisoes in enumerate(hash_table.table):
    if len(colisoes) > 1:
        print(f"Posição {i} teve colisão: {colisoes}")

        total_colisoes += (len(colisoes) - 1)

print(f"\nTotal de colisões encontradas: {total_colisoes}\n")

print("--- Tabela ---\n")

hash_table.imprimir()
```
