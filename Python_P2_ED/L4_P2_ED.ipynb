```python
def questao_1(vetor):
    n = len(vetor)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if vetor[j] < vetor[min_idx]:
                min_idx = j
        vetor[i], vetor[min_idx] = vetor[min_idx], vetor[i]
    return vetor
```

```python
def questao_2(vetor, crescente=True):
    n = len(vetor)
    for i in range(n):
        for j in range(0, n - i - 1):
            if crescente:
                if vetor[j] > vetor[j + 1]:
                    vetor[j], vetor[j + 1] = vetor[j + 1], vetor[j]
            else:
                if vetor[j] < vetor[j + 1]:
                    vetor[j], vetor[j + 1] = vetor[j + 1], vetor[j]
    return vetor
```

```python
def questao_3(vetor):
    if not vetor:
        return None, None
    maior = vetor[0]
    menor = vetor[0]
    for num in vetor:
        if num > maior:
            maior = num
        if num < menor:
            menor = num
    return maior, menor
```

```python
def questao_4(vetor):
    unicos = []
    for num in vetor:
        if num not in unicos:
            unicos.append(num)
    unicos = questao_1(unicos)
    if len(unicos) >= 2:
        return unicos[1]
    return None
```

```python
def questao_5(vetor):
    resultado = []
    for num in vetor:
        if num not in resultado:
            resultado.append(num)
    return resultado
```

```python
def questao_6(vetor):
    vetor_ordenado = questao_2(vetor.copy(), crescente=False)
    pares = 0
    impares = 0
    for num in vetor_ordenado:
        if num % 2 == 0:
            pares += 1
        else:
            impares += 1
    return vetor_ordenado, pares, impares
```

```python
def questao_7(vetor):
    unicos = questao_5(vetor)
    ordenado = questao_2(unicos, crescente=False)
    if len(ordenado) >= 3:
        return ordenado[2]
    return None
```

```python
def questao_8(vetor):
    ordenado = questao_1(vetor.copy())
    meio = len(ordenado) // 2
    return ordenado[meio]
```

```python
entrada = input("Digite os números inteiros separados por espaço: ")
meu_vetor = [int(x) for x in entrada.replace(',', ' ').split()]
print(f"Vetor criado com sucesso: {meu_vetor}")
```

```python
print(f"Vetor original informado: {meu_vetor}\n")

print("--- Questão 1: Select Sort (Crescente) ---")

print(questao_1(meu_vetor.copy()))

print("\n--- Questão 2: Ordenação com Chave ---")

escolha = input("Deseja ordenar de forma (C)rescente ou (D)ecrescente? ").strip().upper()

if escolha == 'C':
    resultado = questao_2(meu_vetor.copy(), crescente=True)
    print(f"Vetor em ordem crescente: {resultado}")
elif escolha == 'D':
    resultado = questao_2(meu_vetor.copy(), crescente=False)
    print(f"Vetor em ordem decrescente: {resultado}")
else:
    print("Opção inválida! Por favor, rode a célula novamente e digite 'C' para crescente ou 'D' para decrescente.")

print("\n--- Questão 3: Maior e Menor ---")
maior, menor = questao_3(meu_vetor)
print(f"Maior valor: {maior} | Menor valor: {menor}")

print("\n--- Questão 4: Segundo Menor (ignorando duplicatas) ---")
print(f"Segundo menor: {questao_4(meu_vetor.copy())}")

print("\n--- Questão 5: Remover Duplicatas ---")
print(questao_5(meu_vetor.copy()))

print("\n--- Questão 6: Decrescente + Contagem de Pares e Ímpares ---")
vetor_ord, pares, impares = questao_6(meu_vetor.copy())
print(f"Vetor Ordenado: {vetor_ord}\nQuantidade de Pares: {pares} | Quantidade de Ímpares: {impares}")

print("\n--- Questão 7: Terceiro Maior (ignorando duplicatas) ---")
print(f"Terceiro maior: {questao_7(meu_vetor.copy())}")

print("\n--- Questão 8: Mediana ---")
print(f"Mediana: {questao_8(meu_vetor.copy())}")
```
