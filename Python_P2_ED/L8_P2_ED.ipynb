```python
import numpy as np

# Dados da matriz
dias = ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta']
semanas = ['S1', 'S2', 'S3', 'S4']

matriz_sono = np.array([
    [8, 6, 7, 8, 6],
    [5, 5, 6, 7, 8],
    [6, 7, 8, 6, 5],
    [7, 5, 9, 7, 4]
])

# 1 e 2
total_por_dia = matriz_sono.sum(axis=0)
dia_maior = dias[np.argmax(total_por_dia)]
dia_menor = dias[np.argmin(total_por_dia)]

# 3 e 4
total_por_semana = matriz_sono.sum(axis=1)
sem_maior = semanas[np.argmax(total_por_semana)]
sem_menor = semanas[np.argmin(total_por_semana)]

# 5 e 6
menor_tempo = matriz_sono.min()
maior_tempo = matriz_sono.max()

# 7 e 8
cont_menor = np.count_nonzero(matriz_sono == menor_tempo)
cont_maior = np.count_nonzero(matriz_sono == maior_tempo)

# 9
total_mes = matriz_sono.sum()
porcentagens = (total_por_semana / total_mes) * 100

print(f'1. Dia com maior descanso: {dia_maior}')
print(f'2. Dia com menor descanso: {dia_menor}')
print(f'3. Semana com maior descanso: {sem_maior}')
print(f'4. Semana com menor descanso: {sem_menor}')
print(f'5. Menor tempo de descanso: {menor_tempo}h')
print(f'6. Maior tempo de descanso: {maior_tempo}h')
print(f'7. Dias com menor tempo ({menor_tempo}h): {cont_menor}')
print(f'8. Dias com maior tempo ({maior_tempo}h): {cont_maior}')
print('9. Porcentagem de descanso por semana:')
for i, p in enumerate(porcentagens):
    print(f'   {semanas[i]}: {p:.2f}%')
```
