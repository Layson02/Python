```python
import numpy as np
# Estruturas Base
class Fila:
    def __init__(self, capacidade):
        self.capacidade = capacidade
        self.inicio = 0
        self.final = -1
        self.numero_elementos = 0
        self.valores = np.empty(self.capacidade, dtype=object)

    def fila_vazia(self):
        return self.numero_elementos == 0

    def fila_cheia(self):
        return self.numero_elementos == self.capacidade

    def enfileirar(self, valor):
        if self.fila_cheia():
            print('A fila está cheia')
            return
        if self.final == self.capacidade - 1:
            self.final = -1
        self.final += 1
        self.valores[self.final] = valor
        self.numero_elementos += 1

    def desenfileirar(self):
        if self.fila_vazia():
            print('A fila já está vazia')
            return
        temp = self.valores[self.inicio]
        self.inicio += 1
        if self.inicio == self.capacidade:
            self.inicio = 0
        self.numero_elementos -= 1
        return temp

    def primeiro(self):
        if self.fila_vazia():
            return -1
        else:
            return self.valores[self.inicio]

    def ultimo(self):
        if self.fila_vazia():
            return -1
        else:
            return self.valores[self.final]

    def tamanho(self):
        return self.numero_elementos

class Pilha:
    def __init__(self):
        self.items = []

    def empilhar(self, item):
        self.items = self.items + [item]

    def desempilhar(self):
        if not self.is_vazia():
            item_removido = self.items[-1]
            self.items = self.items[:-1]
            return item_removido
        else:
            print("A pilha está vazia. Não é possível desempilhar.")

    def topo(self):
        if not self.is_vazia():
            return self.items[-1]
        else:
            print("A pilha está vazia. Não há topo para visualizar.")

    def is_vazia(self):
        return len(self.items) == 0

    def tamanho(self):
        return len(self.items)
```

```python
fila_impressao = Fila(5)

fila_impressao.enfileirar("Relatorio_Financeiro.pdf")
fila_impressao.enfileirar("Folha_Pagamento.pdf")
fila_impressao.enfileirar("Apresentacao_Reuniao.pptx")

print(f"Primeiro documento na fila de impressão: {fila_impressao.primeiro()}")
print(f"Último documento adicionado à fila: {fila_impressao.ultimo()}")
print(f"Total de documentos aguardando impressão: {fila_impressao.tamanho()}")
```

```python
historico_voltar = Pilha()
historico_avancar = Pilha()

historico_voltar.empilhar("google.com")
historico_voltar.empilhar("iesp.edu.br")
historico_voltar.empilhar("github.com")

print(f"Página atual: {historico_voltar.topo()}")

historico_avancar.empilhar(historico_voltar.desempilhar())
print(f"Ação: Voltar página")
print(f"Página atual após voltar: {historico_voltar.topo()}")

historico_voltar.empilhar(historico_avancar.desempilhar())
print(f"Ação: Avançar página")
print(f"Página atual após avançar: {historico_voltar.topo()}")
```

```python
expressao_rpn = "3 4 + 2 *"
pilha_calc = Pilha()

for token in expressao_rpn.split():
    if token.isdigit():
        pilha_calc.empilhar(int(token))
    else:
        v2 = pilha_calc.desempilhar()
        v1 = pilha_calc.desempilhar()

        if token == '+':
            pilha_calc.empilhar(v1 + v2)
        elif token == '-':
            pilha_calc.empilhar(v1 - v2)
        elif token == '*':
            pilha_calc.empilhar(v1 * v2)
        elif token == '/':
            pilha_calc.empilhar(v1 / v2)

print(f"Expressão fornecida: {expressao_rpn}")
print(f"Resultado final do cálculo RPN: {pilha_calc.desempilhar()}")
```

```python
pilha_desfazer = Pilha()
pilha_refazer = Pilha()

pilha_desfazer.empilhar("Digitou 'Olá'")
pilha_desfazer.empilhar("Digitou 'Mundo'")

print(f"Estado do texto atual: {pilha_desfazer.topo()}")

pilha_refazer.empilhar(pilha_desfazer.desempilhar())
print(f"Ação: Desfazer")
print(f"Estado do texto após desfazer: {pilha_desfazer.topo()}")

pilha_desfazer.empilhar(pilha_refazer.desempilhar())
print(f"Ação: Refazer")
print(f"Estado do texto após refazer: {pilha_desfazer.topo()}")
```

```python
expressao = "(2+3)*(8-9)/(7^3)"
pilha_operadores = Pilha()

for char in expressao:
    if char in "+-*/^":
        pilha_operadores.empilhar(char)

print(f"Expressão analisada: {expressao}")
print("Operadores matemáticos encontrados (do último para o primeiro):")

while not pilha_operadores.is_vazia():
    print(f"-> Operador extraído: {pilha_operadores.desempilhar()}")
```

```python
palavra = "radar"
pilha_palindromo = Pilha()

for char in palavra:
    pilha_palindromo.empilhar(char.lower())

palavra_invertida = ""
while not pilha_palindromo.is_vazia():
    palavra_invertida += pilha_palindromo.desempilhar()

if palavra.lower() == palavra_invertida:
    print(f"A palavra original é '{palavra}'.")
    print(f"A palavra invertida pela pilha é '{palavra_invertida}'.")
    print("Conclusão: É um palíndromo.")
else:
    print(f"A palavra original é '{palavra}'.")
    print(f"A palavra invertida pela pilha é '{palavra_invertida}'.")
    print("Conclusão: Não é um palíndromo.")
```
