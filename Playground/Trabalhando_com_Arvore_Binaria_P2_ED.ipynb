# Criando os objetos para a estrutura de árvore binária

A **árvore binária** é uma estrutura de dados hierárquica em que cada nó possui no máximo dois filhos, denominados filho esquerdo e filho direito. Essa estrutura é amplamente utilizada por sua eficiência em operações de busca, inserção e remoção. Uma característica importante é que ela permite organizar os dados de forma que facilite percursos ordenados. Em uma **árvore binária de busca**, por exemplo, os nós à esquerda de um determinado nó contêm valores menores, e os da direita, valores maiores. Essa propriedade torna a árvore binária uma base essencial para algoritmos eficientes e sistemas complexos, como compiladores e bancos de dados.

## Componentes de uma árvore:



* Raiz (root): o nó principal, sem pai.
* Nó (node): elemento da árvore que pode ter filhos.
* Filho (child): nó descendente de outro nó.
* Pai (parent): nó que possui filhos.
* Folha (leaf): nó sem filhos.
* Subárvore: qualquer árvore que faz parte de outra árvore.
* Altura: número de arestas do caminho mais longo da raiz até uma folha.

# Criando os objetos para a estrutura de árvore binária

```python
class No:
    def __init__(self, chave):
        self.chave = chave
        self.esquerda = None
        self.direita = None

class ArvoreBinaria:
    def __init__(self):
        self.raiz = None

    def incluir(self, chave):
        self.raiz = self._incluir(self.raiz, chave)

    def _incluir(self, no, chave):
        if no is None:
            return No(chave)

        if chave < no.chave:
            no.esquerda = self._incluir(no.esquerda, chave)
        elif chave > no.chave:
            no.direita = self._incluir(no.direita, chave)

        return no

    def localizar(self, chave):
        return self._localizar(self.raiz, chave)

    def _localizar(self, no, chave):
        if no is None or no.chave == chave:
            return no
        if chave < no.chave:
            return self._localizar(no.esquerda, chave)
        else:
            return self._localizar(no.direita, chave)

    def visualizar(self):
        self._visualizar(self.raiz)

    def _visualizar(self, no):
        if no is not None:
            self._visualizar(no.esquerda)
            print(no.chave, end=" ")
            self._visualizar(no.direita)

    def pre_order_traversal(self):
        self._pre_order_traversal(self.raiz)
        print()

    def _pre_order_traversal(self, no):
        if no:
            print(no.chave, end=' ')
            self._pre_order_traversal(no.esquerda)
            self._pre_order_traversal(no.direita)

    def in_order_traversal(self):
        self._in_order_traversal(self.raiz)
        print()

    def _in_order_traversal(self, no):
        if no:
            self._in_order_traversal(no.esquerda)
            print(no.chave, end=' ')
            self._in_order_traversal(no.direita)

    def post_order_traversal(self):
        self._post_order_traversal(self.raiz)
        print()

    def _post_order_traversal(self, no):
        if no:
            self._post_order_traversal(no.esquerda)
            self._post_order_traversal(no.direita)
            print(no.chave, end=' ')
```

```python
class No:
    def __init__(self, chave):
        self.chave = chave
        self.esquerda = None
        self.direita = None
        self.altura = 1  # nova propriedade para controlar altura do nó

class ArvoreAVL:
    def __init__(self):
        self.raiz = None

    def incluir(self, chave):
        self.raiz = self._incluir(self.raiz, chave)

    def _incluir(self, no, chave):
        if not no:
            return No(chave)

        if chave < no.chave:
            no.esquerda = self._incluir(no.esquerda, chave)
        elif chave > no.chave:
            no.direita = self._incluir(no.direita, chave)
        else:
            return no  # valor duplicado não é inserido

        # Atualiza altura do nó atual
        no.altura = 1 + max(self._altura(no.esquerda), self._altura(no.direita))

        # Verifica fator de balanceamento
        balanceamento = self._fator_balanceamento(no)

        # Rotação à esquerda
        if balanceamento < -1 and chave > no.direita.chave:
            return self._rotacao_esquerda(no)

        # Rotação à direita
        if balanceamento > 1 and chave < no.esquerda.chave:
            return self._rotacao_direita(no)

        # Rotação dupla esquerda-direita
        if balanceamento > 1 and chave > no.esquerda.chave:
            no.esquerda = self._rotacao_esquerda(no.esquerda)
            return self._rotacao_direita(no)

        # Rotação dupla direita-esquerda
        if balanceamento < -1 and chave < no.direita.chave:
            no.direita = self._rotacao_direita(no.direita)
            return self._rotacao_esquerda(no)

        return no

    def _altura(self, no):
        if not no:
            return 0
        return no.altura

    def _fator_balanceamento(self, no):
        return self._altura(no.esquerda) - self._altura(no.direita)

    def _rotacao_direita(self, y):
        x = y.esquerda
        T2 = x.direita

        # Rotaciona
        x.direita = y
        y.esquerda = T2

        # Atualiza alturas
        y.altura = 1 + max(self._altura(y.esquerda), self._altura(y.direita))
        x.altura = 1 + max(self._altura(x.esquerda), self._altura(x.direita))

        return x

    def _rotacao_esquerda(self, x):
        y = x.direita
        T2 = y.esquerda

        # Rotaciona
        y.esquerda = x
        x.direita = T2

        # Atualiza alturas
        x.altura = 1 + max(self._altura(x.esquerda), self._altura(x.direita))
        y.altura = 1 + max(self._altura(y.esquerda), self._altura(y.direita))

        return y

    def in_order_traversal(self):
        self._in_order_traversal(self.raiz)
        print()

    def _in_order_traversal(self, no):
        if no:
            self._in_order_traversal(no.esquerda)
            print(no.chave, end=' ')
            self._in_order_traversal(no.direita)

    def localizar(self, chave):
        return self._localizar(self.raiz, chave)

    def _localizar(self, no, chave):
        if no is None or no.chave == chave:
            return no  # Retorna o nó encontrado ou None se não existir

        if chave < no.chave:
            return self._localizar(no.esquerda, chave)
        else:
            return self._localizar(no.direita, chave)
```

## Questão 01 - Qual o tempo gasto para construir uma árvore binária de um vetor de 10000 posições?

```python
# Seu código aqui.

import time
import random

vetorQ1 = random.sample(range(1, 10001), 10000)

arvore1 = ArvoreBinaria()
arvore2 = ArvoreAVL()

for v in vetorQ1:
    arvore1.incluir(v)
    arvore2.incluir(v)

valor_procurado = vetorQ1[5000]

inicio = time.perf_counter()
arvore1.localizar(valor_procurado)
fim = time.perf_counter()
print(f"Árvore Binária - Tempo de busca: {fim - inicio:.8f} segundos")

inicio = time.perf_counter()
arvore2.localizar(valor_procurado)
fim = time.perf_counter()
print(f"Árvore AVL     - Tempo de busca: {fim - inicio:.8f} segundos")
```

## Questão 02 - Dada uma lista de inteiros, construa uma árvore binária de busca (BST) inserindo os elementos na ordem em que aparecem na lista.
Exemplo: [10, 5, 15, 3, 7, 12, 18]

```python
# Seu código aqui.

while True:
    entrada = input("Digite os números inteiros (separados por vírgula) para inserir na árvore binária: ")
    try:
        elementos = [int(x.strip()) for x in entrada.split(",") if x.strip()]
        if not elementos:
            raise ValueError
        break
    except ValueError:
        print("Entrada inválida. Exemplo válido: 10, 5, 15, 3, 7")

arvoreQ2 = ArvoreBinaria()

for e in elementos:
    arvoreQ2.incluir(e)

print("Travessia em ordem da árvore binária:")
arvoreQ2.in_order_traversal()
```

## Questão 03 - Dada a árvore construída a partir da lista de 100 elementos, implemente uma função que verifica se um determinado valor existe na árvore.
Entrada: Uma lista de 100 elementos, valor a buscar: 15.

```python
# Seu código aqui.

import random

def existe_na_arvore(arvore, valor):
    return arvore.localizar(valor) is not None

listaQ3 = random.sample(range(1, 201), 100)

arvoreQ3 = ArvoreBinaria()
for numero in listaQ3:
    arvoreQ3.incluir(numero)

print("Lista Q3 ordenada (menor -> maior):")
arvoreQ3.in_order_traversal()

valor_busca = 15

if existe_na_arvore(arvoreQ3, valor_busca):
    print(f"O valor {valor_busca} existe na árvore.")
else:
    print(f"O valor {valor_busca} não existe na árvore.")
```

## Questão 04 - Dada uma lista de 1000 números inteiros, contrua uma árvore a partir desta lista e em seguida responda:

* Verifique se existem algum desses elementos na estrutura da árvore binária: [95, 120, 87, 75, 60].


```python
# Seu código aqui.

import random

listaQ4 = random.sample(range(1, 5001), 1000)

arvoreQ4 = ArvoreBinaria()
for numero in listaQ4:
    arvoreQ4.incluir(numero)

print("Lista Q4 ordenada (menor -> maior):")
arvoreQ4.in_order_traversal()

valores_verificar = [95, 120, 87, 75, 60]

print("Resultado da verificação na árvore binária:")
for valor in valores_verificar:
    encontrado = arvoreQ4.localizar(valor) is not None
    status = "encontrado" if encontrado else "não encontrado"
    print(f"Valor {valor}: {status}")
```
